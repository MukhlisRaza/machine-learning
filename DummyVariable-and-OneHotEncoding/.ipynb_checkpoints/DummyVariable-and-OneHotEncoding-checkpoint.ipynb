```python
import pandas as pd
```

```python
dataset = pd.read_csv("towns_dataset.csv") #Loading csv file
dataset
```

```python
dummies_variable = pd.get_dummies(dataset.Town)
dummies_variable
```

```python
dummies_dataset = pd.concat([dataset,dummies_variable],axis="columns") #concatenate the dummies variable with dataset
dummies_dataset
```

# Dummies variable Trap

```python
dummies_dataset.drop(["Town","satellite town RWP"],axis="columns",inplace=True)
dummies_dataset
```

```python
X = dummies_dataset.drop("Price",axis="columns") # we also drop the price column becuause it a targeted columns
X
```

```python
y = dummies_dataset.Price
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y) # We train our model
```

```python
# Build predictor function predict price of the home

model.predict([[3400,0,1]]) #predicting in the Behria Town
```

# one hot encoding
